```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("dataCrawlWeb.json")
```

```python
df[["titleOfJobs", "companyName", "countryOfcompany"]].head()
```

```python
country = df.groupby("countryOfcompany")["titleOfJobs"]
```

```python
plt.figure(figsize=(15,10))
country.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Wines")
plt.show()
```

```python
temp1 = [i for i in df.technologyLlist]
temp2 = []
for i in temp1:
    temp2.append(" ".join(i))
temp2 = " ".join(temp2)
# Start with one review:
text = temp2

# Create and generate a word cloud image:
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
workingDate = df.groupby("workingDate")["titleOfJobs"].agg(["count"])
labels = ['workingDate', 'Monday - Saturday', 'Unkown']
sizes = [i for i in workingDate["count"]]
explode = (0, 0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
