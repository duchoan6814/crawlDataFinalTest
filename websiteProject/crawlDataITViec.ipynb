```python
from selenium import webdriver
import time
import json, codecs
driver = webdriver.Chrome()
driver.get('https://itviec.com/it-jobs')
```

```python
#function

# Lấy ra tổng các jobs mà web nó hiện thị ở đầu trang web
def getTotalJobs():
    textTitleOfAllJobs = driver.find_element_by_css_selector('div.col-xs-12>h1').text
    listTextInside = textTitleOfAllJobs.split(' ')
    return listTextInside[0]
# Format lại số jobs lấy ra để chuyển đổi qua int
def formatTotalJobs(totalJobs):
    a = totalJobs.split(',')
    a = "".join(a)
    return a
# Lấy tất cả những jobs đang hiển thị trên page
def getJobsShowedInMonitor(numberOfJobs):
    jobsVisible = []
    while (int(len(driver.find_elements_by_css_selector('div.job'))) <= numberOfJobs):
        driver.find_element_by_css_selector('a.more-jobs-link').click()
        jobsVisible = driver.find_elements_by_css_selector('div.col-xs-12 div.job')
        time.sleep(1)
    return jobsVisible
# Lấy Danh sách link từ list jobs đã lấy được
def getLinkTotalJobs(listJobs):
    listLink = []
    for i in listJobs:
        try:
            listLink.append(i.find_element_by_css_selector('h2.title>a').get_attribute('href'))
        except:
            pass
    return listLink
# Lấy list chi tiết jobs
def getListDetailJobs(listLinkJobs):

    ListDetailJobs = []
    loading = 0

    for i in listLinkJobs:
        detailJobs = {}
        driver.get(i)
        try:
            titleOfJobs = driver.find_element_by_css_selector('div.job_info>h1.job_title').text
        except:
            titleOfJobs = "Unknown"
        try:
            address = driver.find_element_by_css_selector('div.address__full-address>span').text
        except:
            address = "Unknown"
        try:
            technologyLlist = []
            for i in driver.find_elements_by_css_selector('div.job-detail div.tag-list>a'):
                technologyLlist.append(i.text)
        except:
            TechnologyLlist = []
        try:
            companyName = driver.find_element_by_css_selector('div.employer-info > h3.name > a').text
        except:
            companyName = "Unknown"
        try:
            countryOfCompany = driver.find_element_by_css_selector('div.country-icon > span.country-name').text
        except:
            countryOfCompany = "Unknown"
        try:
            salary = driver.find_element_by_css_selector('div.salary > span.salary-text').text
        except:
            salary = "Unknown"
        try:
            top3ReasonToJoinUs = []
            for i in driver.find_elements_by_css_selector('div.top-3-reasons li'):
                top3ReasonToJoinUs.append(i.text)
        except:
            top3ReasonToJoinUs = []
        try:
            theJob = []
            for i in driver.find_elements_by_css_selector('div.job_description li'):
                theJob.append(i.text)
        except:
            theJob = []
        try:
            skillsExperience = []
            for i in driver.find_elements_by_css_selector('div.skills_experience li'):
                skillsExperience.append(i.text)
        except:
            skillsExperience = []
        try:
            loveWorkingHere = []
            for i in driver.find_elements_by_css_selector('div.love_working_here li'):
                loveWorkingHere.append(i.text)
        except:
            loveWorkingHere = []
        try:
            workingDate = driver.find_element_by_css_selector('div.working-date span').text
        except:
            workingDate = "Unknown"
        detailJobs["titleOfJobs"] = titleOfJobs
        detailJobs["address"] = address
        detailJobs["technologyLlist"] = technologyLlist
        detailJobs["companyName"] = companyName
        detailJobs["countryOfcompany"] = countryOfCompany
        detailJobs["salary"] = salary
        detailJobs["workingDate"] = workingDate
        detailJobs["top3ReasonToJoinUs"] = top3ReasonToJoinUs
        detailJobs["theJob"] = theJob
        detailJobs["skillsExperience"] = skillsExperience
        detailJobs["loveWorkingHere"] = loveWorkingHere
        ListDetailJobs.append(detailJobs)
        loading = loading + 1
        print(str((loading/len(listLinkJobs)) * 100) + "%")

    print("done")

    return ListDetailJobs
```

```python
#main
if __name__ == "__main__":
    listLinkJobs = getLinkTotalJobs(getJobsShowedInMonitor(int(formatTotalJobs(getTotalJobs()))))
    # listLinkJobs = getLinkTotalJobs(getJobsShowedInMonitor(40))
    listDetailJobs = getListDetailJobs(listLinkJobs)
    with open('dataCrawlWeb.json', "wb") as outfile:
        json.dump(listDetailJobs, codecs.getwriter('utf-8')(outfile), ensure_ascii = False)
```
