```python
from selenium import webdriver
import json, codecs
import datetime
import time
driver = webdriver.Chrome()
driver.get("https://www.facebook.com/groups/iuh2020")

def handleDate(date):
    return datetime.datetime(int(date.split(", ")[2].split(" ")[0]), int(date.split(", ")[1].split(" ")[2]), int(date.split(", ")[1].split(" ")[0]), int(date.split(", ")[2].split(" ")[2].split(":")[0]), int(date.split(", ")[2].split(" ")[2].split(":")[1]))

def checkStr(str):
    if str.find("K") >= 0:
        return True
    return False

# lấy list link post
listLink = []
temp = []
position = 12000
while len(temp) <= 500:
    driver.execute_script("window.scrollTo(0,"+str(position)+" )")
    temp = driver.find_elements_by_css_selector('div._1dwg._1w_m._q7o')
    position+=12000
for i in temp:
    listLink.append(i.find_element_by_css_selector('a._5pcq').get_attribute('href'))
    
# main
listPost = []
loading = 0
for i in listLink:
    driver.get(i)
    tempDic = {}
    try:
        nameOfPersonPost = driver.find_element_by_css_selector("span._39_n").get_attribute('title')
    except:
        try:
            nameOfPersonPost = driver.find_element_by_css_selector('a.profileLink').text
        except:
            nameOfPersonPost = "None"
    try:
        contentOfPost = driver.find_element_by_css_selector('div._5pbx.userContent._3ds9._3576').text
    except:
        try:
            contentOfPost = driver.find_element_by_css_selector('span._5z6m > span._4a6n').text
        except:
            try:
                contentOfPost = driver.find_element_by_css_selector('div._5pbx.userContent._3576 > p').text
            except:
                contentOfPost = "None"
    try:
        dateTimeOfPost = handleDate(driver.find_element_by_css_selector("a._5pcq > abbr").get_attribute('title'))
    except:
        dateTimeOfPost = "None"
    try:
        InteractiveOfPost = driver.find_element_by_css_selector('span._81hb').text.split(",")[0] + driver.find_element_by_css_selector('span._81hb').text.split(",")[1].split("K")[0] + "00"
    except:
        try:
            InteractiveOfPost = driver.find_element_by_css_selector('span._81hb').text
        except:
            InteractiveOfPost = "0"
    try: 
        commentsOfPost = driver.find_element_by_css_selector('a._3hg-._42ft').text.split(" ")[0]
    except:
        commentsOfPost = "0"
#     lấy các comment
    try:
        driver.find_element_by_css_selector('a._3hg-._42ft').click()
    except:
        pass
    time.sleep(1)
    while True:
        try:
            driver.find_element_by_css_selector('a._4sxc._42ft').click()
        except:
            break
    listComment = driver.find_elements_by_css_selector('div._72vr')
    listComments = []
    for i in listComment:
        dicTemp = {}
        try:
            userName = i.find_element_by_css_selector('a._6qw4').text
        except:
            userName = i.find_element_by_css_selector('span._6qw4').text
        try:
            contentComment = i.find_element_by_css_selector('span._3l3x').text
        except:
            contentComment = "None"
        dicTemp["UserName"] = userName
        dicTemp["ContentComment"] = contentComment
        listComments.append(dicTemp)
    
    tempDic["Tên Người Post"] = nameOfPersonPost
    tempDic["Nội dung Post"] = contentOfPost
    tempDic["Ngày Đăng"] = str(dateTimeOfPost)
    tempDic["Số cảm xúc"] = int(InteractiveOfPost)
    tempDic["Số bình luận"] = int(commentsOfPost)
    tempDic["Các bình luận"] = listComments
    listPost.append(tempDic)
    loading = loading + 1
    print(str((loading/len(listLink)) * 100) + "%")
with open('data.json', "wb") as outfile:
    json.dump(listPost, codecs.getwriter('utf-8')(outfile), ensure_ascii = False)
print("Done")
    
```
